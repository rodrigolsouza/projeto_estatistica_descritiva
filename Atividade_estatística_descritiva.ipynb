## 3. Análise de Quartis e Outliers

Os quartis dividem um dataset ordenado em quatro partes iguais (Q1: 25%, Q2: 50% - mediana, Q3: 75%). Eles são cruciais para entender a dispersão e identificar possíveis outliers. Um outlier é um ponto de dado que se desvia significativamente de outras observações. Podemos identificá-los usando o Intervalo Interquartil (IQR), que é a diferença entre Q3 e Q1, e estabelecendo limites inferior e superior.

### Exercício 3.1: Cálculo Manual de Quartis e Identificação de Outliers

Para a coluna 'total_bill', calcule manualmente (sem usar `describe()` diretamente para os quartis, mas pode usar `quantile()`) os valores de Q1, Q2 (mediana), Q3 e o Intervalo Interquartil (IQR). Em seguida, calcule o Limite Inferior e Superior para identificação de outliers (Limite Inferior = Q1 - 1.5 * IQR; Limite Superior = Q3 + 1.5 * IQR). Por fim, mostre quantos outliers existem na coluna 'total_bill'.

```python
# Cálculo do Q1:
Q1 = df['total_bill'].quantile(0.25)
print(f"Q1 (25º percentil) da coluna 'total_bill': {Q1}")

Q2 = df['total_bill'].quantile(0.5)
print(f"Q2 (50º percentil) da coluna 'total_bill': {Q2}")
# Cálculo do Q3:
Q3 = df['total_bill'].quantile(0.75)
print(f"Q3 (75º percentil) da coluna 'total_bill': {Q3}")

# Cálculo do IQR:
IQR = Q3 - Q1
print(f"IQR (Intervalo entre os Quartis) da coluna 'total_bill': {IQR}")
# Cálculo dos limites inferior e superior para detecção de outliers:
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Limite inferior para detecção de outliers: {lower_bound}")
print(f"Limite superior para detecção de outliers: {upper_bound}")
#Identificação de outliers na coluna 'total_bill':
outliers = df[(df['total_bill'] < lower_bound) | (df['total_bill'] > upper_bound)]
print(f"Número de outliers na coluna 'total_bill': {len(outliers)}")
```

## 4. Assimetria (Skewness) e Curtose

A assimetria (skewness) mede o grau de inclinação de uma distribuição de probabilidade. Uma distribuição simétrica tem skewness zero. Valores positivos indicam uma cauda direita mais longa, enquanto valores negativos indicam uma cauda esquerda mais longa.

A curtose (kurtosis) mede o 'achatamento' ou 'pico' de uma distribuição. Uma curtose alta significa mais valores nas caudas e um pico mais nítido (leptocúrtica), enquanto uma curtose baixa indica menos valores nas caudas e um pico mais plano (platocúrtica). Uma distribuição normal tem curtose próxima de 3 (ou 0, se for calculada a curtose em excesso).

### Exercício 4.1: Cálculo e Interpretação de Assimetria e Curtose

Calcule a assimetria (skewness) e a curtose para as colunas numéricas ('total_bill', 'tip', 'size').

**Pergunta Dissertativa:** Com base nos valores calculados, descreva se a distribuição de cada variável numérica é simétrica, enviesada para a esquerda ou para a direita, e se ela é leptocúrtica, mesocúrtica ou platocúrtica. O que isso implica sobre a concentração dos dados?

```python
# DIGITE SEU CÓDIGO AQUI

import pandas as pd
import seaborn as sns # Apenas para carregar o dataset 'tips' se você não tiver o arquivo

# Carregando os dados (exemplo usando o dataset padrão do seaborn)
df = sns.load_dataset('tips')

# Selecionando as colunas numéricas do exercício
colunas = ['total_bill', 'tip', 'size']

# Calculando Assimetria (Skewness)
assimetria = df[colunas].skew()

# Calculando Curtose (Kurtosis)
# Nota: O pandas calcula a Curtose em Excesso (onde a Normal é 0)
curtose = df[colunas].kurt()

# Exibindo os resultados de forma organizada
print("=== Análise Estatística ===")
for col in colunas:
    print(f"\nColuna: {col}")
    print(f"Assimetria: {assimetria[col]:.4f}")
    print(f"Curtose: {curtose[col]:.4f}")
```

A visualização de dados é uma parte essencial da EDA. Gráficos como histogramas e boxplots nos permitem entender rapidamente a distribuição, dispersão e a presença de outliers, muitas vezes de forma mais intuitiva do que apenas com números.

### Exercício 6.1: Histogramas com KDE

Crie histogramas com a Estimativa de Densidade Kernel (KDE) para as colunas numéricas: 'total_bill', 'tip' e 'size'. Os histogramas mostram a frequência de valores em intervalos (bins), e o KDE suaviza essa distribuição, mostrando a densidade de probabilidade.

```python
# DIGITE SEU CÓDIGO AQUI

# 1. Carregando o dataset (exemplo com o dataset 'tips')
df = sns.load_dataset('tips')

# 2. Configurando o estilo visual
sns.set_theme(style="whitegrid")

# 3. Criando a figura com subplots (1 linha e 3 colunas)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Lista de colunas para facilitar o loop
colunas = ['total_bill', 'tip', 'size']
cores = ['blue', 'green', 'orange']

# 4. Gerando os gráficos
for i, col in enumerate(colunas):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i], color=cores[i], bins=20)
    axes[i].set_title(f'Distribuição de {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequência / Densidade')

# Ajustando o layout para não sobrepor
plt.tight_layout()
plt.show()
```

## 7. Normalização de Dados

A normalização é um processo crucial no pré-processamento de dados, especialmente quando algoritmos de Machine Learning são sensíveis à escala das features (por exemplo, modelos baseados em distância como KMeans, SVMs com kernels, ou Redes Neurais). O objetivo é ajustar os valores das features para uma escala comum, sem distorcer as diferenças nas faixas de valores ou perder informações. Existem duas técnicas comuns:

1.  **Min-Max Scaling (Normalização)**: Escala os dados para um intervalo fixo, geralmente entre 0 e 1. A fórmula é: `X_novo = (X - X_min) / (X_max - X_min)`.
2.  **StandardScaler (Padronização - Z-score)**: Transforma os dados para ter média 0 e desvio padrão 1. A fórmula é: `X_novo = (X - média) / desvio_padrão`.

### Exercício 7.1: Aplicação de Normalização Min-Max e StandardScaler

Utilize as colunas numéricas ('total_bill', 'tip', 'size') para aplicar dois métodos de normalização:

1.  **Min-Max Scaling**: Normalize estas colunas para o intervalo [0, 1]. Armazene os resultados em um novo DataFrame ou adicione colunas normalizadas ao DataFrame `df` (ex: `total_bill_minmax`).
2.  **StandardScaler (Z-score)**: Padronize as mesmas colunas para ter média 0 e desvio padrão 1. Armazene os resultados em um novo DataFrame ou adicione colunas padronizadas ao DataFrame `df` (ex: `total_bill_zscore`).

Lembre-se de importar `MinMaxScaler` e `StandardScaler` do `sklearn.preprocessing`.

```python
# DIGITE SEU CÓDIGO AQUI
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Selecionando as colunas numéricas para normalização
df = sns.load_dataset('tips')

cols = ['total_bill', 'tip', 'size']

# 1. Min-Max Scaling
min_max=MinMaxScaler()
df[['total_bill_minmax', 'tip_minmax', 'size_minmax']] = min_max.fit_transform(df[cols])

# 2. StandardScaler (Z-score)
std_scale = StandardScaler()
df[['total_bill_zscore', 'tip_zscore', 'size_zscore']] = std_scale.fit_transform(df[cols])

# Verificando as colunas Min-Max (esperado: min=0, max=1)
print("--- Verificação Min-Max ---")
print(df[['total_bill_minmax', 'tip_minmax', 'size_minmax']].describe().loc[['min', 'max']])

print("\n--- Verificação Z-score ---")
# Verificando as colunas Z-score (esperado: mean=0, std=1)
print(df[['total_bill_zscore', 'tip_zscore', 'size_zscore']].describe().loc[['mean', 'std']])

```

### Análise do Aluno:
[Compare os valores originais com os valores normalizados por Min-Max e padronizados por StandardScaler. Observe as diferenças na escala e no range dos dados. Quando seria mais apropriado usar um método em vez do outro?]
1. Use Min-Max Scaling quando:
Algoritmos que exigem limites fixos: Como Redes Neurais (onde as funções de ativação esperam entradas entre 0 e 1).

Processamento de Imagem: Onde os pixels são naturalmente representados em escalas fixas (0-255).

Distância Importa (sem Outliers): Algoritmos como K-Nearest Neighbors (KNN), desde que seus dados não tenham valores muito extremos que possam distorcer toda a escala.

2. Use StandardScaler quando:
Algoritmos que assumem Distribuição Normal: Muitos modelos lineares (Regressão Linear, Logística) e o SVM funcionam melhor se os dados estiverem centrados na média.

Análise de Componentes Principais (PCA): O PCA busca maximizar a variância; se você não padronizar, as variáveis com números maiores dominarão os componentes falsamente.

Presença de Outliers: O Z-score é mais "robusto". Se você tiver uma conta de $500 no seu dataset, o Min-Max jogaria todas as outras contas ($10, $20) para valores minúsculos como 0.001. O Z-score apenas daria uma nota alta (ex: 15.0) para essa conta cara, mantendo as outras com uma distribuição legível.
